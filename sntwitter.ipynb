```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

tweets_list = []

keyword = input('input keyword : ')
start_date = input('start_date (yyyy-mm-dd) : ')
end_date = input('end_date (yyyy-mm-dd) : ')
max = int(input('maximum tweet number : '))
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(keyword +' since:' + start_date + ' until:' + end_date).get_items()):
    if i > max:
        break
    tweets_list.append([tweet.date, tweet.content, tweet.username, tweet.likeCount])
    
  
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Text','Username','Like'])
tweets_df['Datetime'] = tweets_df['Datetime'].apply(lambda a: pd.to_datetime(a).date())
tweets_df.to_excel(keyword + '_' + start_date +'_' + end_date + 'tweetdata.xlsx')


# ì…ë ¥ìœ¼ë¡œ 2021-01-01ë¶€í„° 2021-12-31ì„ ë°›ìœ¼ë©´ 2021-12-30ì¼ ìë£Œë¶€í„° ì—­ìˆœìœ¼ë¡œ ì €ì¥ë¨
# BTS 2022-01-01 2022-03-01 ì„ ì••ë ¥ìœ¼ë¡œ ë°›ìœ¼ë©´
# 2022-02-28 11:59 -> 2022-02-28 11:58 .....-> 2022-01-01 00:00 ìˆœìœ¼ë¡œ ì €ì¥ë¨

# csv íŒŒì¼ë¡œ ì €ì¥í•˜ë ¤ë©´ ë§¨ ë°‘ ì½”ë“œë§Œ ë°”ê¾¸ë©´ë¨
#ì½”ë©ì—ì„œëŠ” ì•ˆëŒì•„ê°€ë‹ˆê¹Œ ì£¼í”¼í„°ë¡œ ëŒë ¤
```
